```python
# finding total gender columns 
population_df.replace("...", 0)

for column in population_df.columns:
    population_df[column] = population_df[column].apply(lambda value: int(value) if (isinstance(value, str) and value.isdigit()) else value)

male = population_df.iloc[:,population_df.columns.str.startswith('Male persons age')]
female = population_df.iloc[:,population_df.columns.str.startswith('Female persons age')]
blackmale = population_df.iloc[:,population_df.columns.str.startswith('Black male persons age')]
blackfemale = population_df.iloc[:,population_df.columns.str.startswith('Black female persons age')]


male_cols = population_df.iloc[:,population_df.columns.str.startswith('Male persons age')]
female_cols = population_df.iloc[:,population_df.columns.str.startswith('Female persons age')]
blackmale_cols = population_df.iloc[:,population_df.columns.str.startswith('Black male persons age')]  
blackfemale_cols = population_df.iloc[:,population_df.columns.str.startswith('Black female persons age')]

population_df["Total Male Persons"] = male.sum(axis=1)
population_df["Total Female Persons"] = female.sum(axis=1)
population_df["Total Black Male Persons"] = blackmale.sum(axis=1)
population_df["Total Black Female Persons"] = blackfemale.sum(axis=1)

population_df2 = population_df.drop(list(female_cols) + list(male_cols) + list(blackmale_cols) + list(blackfemale_cols), axis=1)
#population_df3 = population_df2.drop(list(male_cols), axis=1)

population_df2[population_df2.get('Total Black Male Persons') != '........................']
```
