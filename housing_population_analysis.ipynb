```python
# Imports relevant CSVs
housing_df = pd.read_csv("SD1970_housing.csv")
population_df = pd.read_csv("SD1970_population.csv")
```

```python
population_df.replace("...", 0)

for column in population_df.columns:
    population_df[column] = population_df[column].apply(lambda value: int(value) if (isinstance(value, str) and value.isdigit()) else value)

male = population_df.iloc[:, 4:26]
male_cols = population_df.iloc[:, 4:26].columns
female = population_df.iloc[:, 26:48]
female_cols = population_df.iloc[:, 26:48].columns
    
population_df["Total Male Persons"] = male.sum(axis=1)
population_df = population_df.drop(list(male_cols), axis=1)

population_df["Total Female Persons"] = female.sum(axis=1)
population_df = population_df.drop(list(female_cols), axis=1)

population_df.head()
```
